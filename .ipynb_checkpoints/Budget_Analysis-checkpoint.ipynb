## **Budget Analysis**

#### Import Statements

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Read Excel File

```python
data = pd.read_excel("C:\\Users\\f-eng\\OneDrive\\Documents\\Python\\Projects\\Budget_Program\\Accounting_2020.xlsx", engine="openpyxl")
column_names = data.columns
months = ("JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT","NOV","DEC")
#print(data.to_string())
```

#### Filter by Credit Only

```python
credit = data[data.CREDIT.notnull()]
credit.reset_index(inplace=True)
credit.head()
```

#### Filter by Debit Only

```python
debit = data[data["DEBIT"].notnull()]
debit.reset_index(inplace=True)
debit.head()
```

#### Get Monthly Income/Spending

```python
monthly_income = credit.groupby('MONTH')["CREDIT"].sum()
average_monthly_income = [round(monthly_income.mean(),2)]
print(monthly_income)
print(average_monthly_income)
monthly_spending = debit.groupby("MONTH")["DEBIT"].sum()
average_monthly_spending = [round(monthly_spending.mean(), 2)]
print(monthly_spending)
print(average_monthly_spending)
gain = round((monthly_income.sum() - monthly_spending.sum()), 2)
print(gain)
estimated_gain = round((average_monthly_income[0] - average_monthly_spending[0])*12, 2)
print(estimated_gain)
total_spending = monthly_spending.sum()
print(total_spending)
```

```python
plt.figure(figsize=(32,8))
plt.title("Monthly Income/Spending in 2020", fontdict={"fontweight":"bold", "fontsize":30})
plt.plot(months, monthly_income, label="Income", linewidth=6)
# plt.plot(months, [average_monthly_income]*12, label="Average Income")
# plt.plot(months, [average_monthly_spending]*12, label="Average Spending")
plt.plot(months, monthly_spending, label="Spending", linewidth=6)
plt.ylabel("Income/Spending $", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(loc=2, prop={'size':30})
```

```python
plt.figure(figsize=(32,8))
plt.title("Monthly Spending and Average Spending in 2020", fontdict={"fontweight":"bold", "fontsize":30})
plt.plot(months, [average_monthly_spending]*12, linestyle="--", label="Average Spending", linewidth=6)
plt.plot(months, monthly_spending, marker=".", markersize=30, label="Spending", linewidth=6)
plt.ylabel("Spending $", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(prop={'size':30})
```

```python

fig, ax = plt.subplots(figsize=(28,8))
bar = ax.bar(months, monthly_spending)
ax.plot(months, [average_monthly_spending]*12,"r--", label="Average Spending", linewidth=2)
ax.tick_params(axis='both', labelsize=20)
ax.set_ylim([0,10000])
ax.set_title("Monthly Spending and Average Spending in 2020", fontdict={"fontweight":"bold", "fontsize":30})
ax.legend(loc='best', fontsize=20)
def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0.5, 'right': 0.57, 'left': 0.43}  # x_txt = x + w*off

    for rect in rects:
        height = round(rect.get_height(), 2)
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.08*height,
                '{}'.format(height), ha=ha[xpos], va='bottom', size=15)
autolabel(bar)
```
